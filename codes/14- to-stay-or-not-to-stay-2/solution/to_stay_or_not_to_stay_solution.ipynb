## Part Two: Analysis

### Question 1

```python
# TODO: find the top 3 countries
indices = (df['2020'].sort_values()[-3:].index).tolist()
top_countries = df.loc[indices]['Country'].tolist() [::-1]
print(top_countries)
```

### Question 2

```python
# TODO: find iran_mean
iran_mean = df[df['Country'] == "Iran (Islamic Republic of)"].drop(columns=['Country']).astype(np.float64).mean(axis=1).item()
print(iran_mean)
```

### Question 3

```python
# TODO: find highest_growth
highest_growth = df.loc[(df['2020'] - df['1990']).sort_values()[-1:].index]['Country'].item()
print(highest_growth)
```
