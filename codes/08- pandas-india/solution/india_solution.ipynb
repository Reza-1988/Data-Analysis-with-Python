```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('food.csv')
df.head()
```

```python
# Count the number of each flavor profile
counts = df['flavor_profile'].value_counts()
counts
```

```python
# Calculate the ratio of spicy foods
spicy_ratio = counts['spicy'] / counts.sum()
spicy_ratio
```

```python
def calculate_full_time(row):
    # Add the prep time and cook time
    return row['prep_time'] + row['cook_time']
```

```python
# Calculate the full time for each row
df['full_time'] = df.apply(calculate_full_time, axis=1)
df.head()
# df['full_time'] = df.apply(lambda row: row['prep_time'] + row['cook_time'], axis=1) Alternative way 
```

```python
# Find the top 10 longest cooking dishes
df_most_time_consuming = df.sort_values(by=['full_time','name'], ascending=[False,True])[:10]
df_most_time_consuming
```

```python
# drop full time
df_most_time_consuming.drop('full_time' ,axis=1 ,inplace=True)
```
